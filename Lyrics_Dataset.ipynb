```python

def create_genre_df(playlists, sp, genre=''):
    data = {
        'Artist Name': [],
        'Track Name': [],
        'Album Name': [],
        'Artist URI': [],
        'Album URI': [],
        'Genre': [],
    }

    for playlist_uri in playlists:
        try:
            results = sp.playlist(playlist_uri)

            if results is not None and 'tracks' in results:
                for track in results['tracks']['items']:
                    if 'track' in track and 'artists' in track['track'] and track['track']['artists']:
                        artist = track['track']['artists'][0]['name']
                        track_name = track['track']['name']
                        album_name = track['track']['album']['name']
                        artist_uri = track['track']['artists'][0]['uri']
                        album_uri = track['track']['album']['uri']

                        data['Artist Name'].append(artist)
                        data['Track Name'].append(track_name)
                        data['Album Name'].append(album_name)
                        data['Artist URI'].append(artist_uri)
                        data['Album URI'].append(album_uri)
                        data['Genre'].append(genre)
        except Exception as e:
            print(f"An error occurred while processing playlist {playlist_uri}: {str(e)}")

    df = pd.DataFrame(data)
    return df
```

```python
def get_lyrics_df(songs, artists, max_retries=1):
    lyrics_data = []
    for song, artist in zip(songs, artists):
        retry_count = 0
        while retry_count < max_retries:
            try:
                song_info = genius.search_song(title=song, artist=artist)
                if song_info is not None and hasattr(song_info, 'lyrics'):
                    lyrics = song_info.lyrics
                    lyrics_data.append({'Track_Name': song, 'Artist_Name': artist, 'Lyrics': lyrics})
                    break
            except:
                pass
            retry_count += 1

    lyrics_df = pd.DataFrame(lyrics_data)

    return lyrics_df
```

```python
df = get_lyrics_df(songslist, artistslist)
df
```
